```python
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import create_react_agent
from langchain_core.messages import HumanMessage
from langchain_core.tools import tool
# Set your OpenAI API key
from google.colab import userdata
OPENAI_API_KEY = userdata.get("OpenAI-key")

# Define a tool
@tool
def get_weather(location: str):
    """Get the current weather in a specific location."""
    return f"Weather in {location}: Sunny, 75°F"

# Initialize model and tools
model = ChatOpenAI(openai_api_key=OPENAI_API_KEY, model="gpt-4o", temperature=0)
tools = [get_weather]

# Create the agent
agent = create_react_agent(model, tools)

# Run the agent
inputs = {"messages": [HumanMessage(content="What is the weather in San Francisco?")]}
result = agent.invoke(inputs)

# Print result
for msg in result["messages"]:
    msg.pretty_print()
```
